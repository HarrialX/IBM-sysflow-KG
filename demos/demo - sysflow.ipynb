### back-up: num of CWE from BRON and by our method

```python
import numpy as np
from tqdm import tqdm

def dcg_at_k(rel, k):
    """
    calculate discounted cumulative gain (dcg)
    rel: list, element is positive real values, can be binary
    """
    rel = np.asfarray(rel)[:k]
    dcg = np.sum((2 ** rel - 1) / np.log2(np.arange(2, rel.size + 2)))
    return dcg


def ndcg_at_k(rel, k):
    """
    calculate normalized discounted cumulative gain (ndcg)
    rel: list, element is positive real values, can be binary
    """
    idcg = dcg_at_k(sorted(rel, reverse=True), k)
    if not idcg:
        return 0.
    return dcg_at_k(rel, k) / idcg


def hit_at_k(hit, k):
    """
    calculate Hit@k
    hit: list, element is binary (0 / 1)
    """
    hit = np.asarray(hit)[:k]
    return int(np.any(hit[:]==1))

    
def cwe_by_bron(techs: list):
    capecs = set()
    for h, r, t in sf.factset[sf.rel_dict['technique:attack-pattern']]:
        assert h in sf.entset['technique']
        assert t in sf.entset['attack-pattern']
        for tech in techs:
            
            if tech in h:
                capecs.add(t)

    cwes = set()
    for h, r, t in sf.factset[sf.rel_dict['attack-pattern:weakness']]:
        assert h in sf.entset['attack-pattern']
        assert t in sf.entset['weakness']
        if h in capecs:
            cwes.add(t)
    return cwes


def all_bron_cwe():
    capecs = set()
    for h, r, t in sf.factset[sf.rel_dict['technique:attack-pattern']]:
        capecs.add(t)

    cwes = set()
    for h, r, t in sf.factset[sf.rel_dict['attack-pattern:weakness']]:
        if h in capecs:
            cwes.add(t)
    return cwes

all_bron_cwe = all_bron_cwe()
# bron_cwe = cwe_by_bron(['T1046'])
# thresholds = [0, 0.04, 0.08]
n_cves = [100, 200, 300]
top = 5


def metric(multiview: bool, filtering: bool):
    precision = []
    recall = []
#     hit = []
#     ndcg = []
#     mrr = []
    
    if filtering:
        cdd_cve = sf.cdd_cve_by_kw('product', 'Linux')
    else:
        cdd_cve = None
        
    # for thre in thresholds:
    for thre in n_cves:
#         cnt, m1, m2, m3, m4, m5 = 0, 0, 0, 0, 0, 0
        cnt, m1, m2 = 0, 0, 0
        for ttp in tqdm(sf.tech_scores.keys(), desc='threshold/n_cve %s' % thre):
            bron_cwe = cwe_by_bron([ttp])
            if len(bron_cwe)==0:
                continue
    #         thre_cves, thre_scores = sf.ttp_cve_link(thre=thre, tech = ttp, multiview=multiview)
            thre_cves, thre_scores = sf.ttp_cve_link(
                n_cve=thre, tech = ttp, multiview=multiview, cdd_cve=cdd_cve, filtering=filtering)
            cwe_list_sort, cwe_score_sort = sf.cwe_cve_link(thre_cves, thre_scores)
            link_cwe = {}
            for i in range(len(cwe_list_sort)):
                link_cwe[cwe_list_sort[i]] = cwe_score_sort[i]
            link_cwe = [k for k, v in sorted(link_cwe.items(), key=lambda item: item[1], reverse=True) 
                        if k in all_bron_cwe]
            if len(link_cwe)==0:
                continue

    #         print('TTP', ttp)
    #         print("threshold:", thre)
    #         print('bron_cwe', bron_cwe)
    #         print('linked_cwe', link_cwe)
    #         print('precision %.2f' % (len(set(link_cwe) & set(bron_cwe)) / len(link_cwe)))
    #         print('recall %.2f' % (len(set(link_cwe) & set(bron_cwe)) / len(bron_cwe)))
    #         print('-'*20)
            cnt += 1
            m1 += len(set(link_cwe[:top]) & set(bron_cwe)) / len(link_cwe[:top])
            m2 += len(set(link_cwe[:top]) & set(bron_cwe)) / len(bron_cwe)

#             binary_hit = [1 if cwe in bron_cwe else 0 for cwe in link_cwe]
#             m3 += hit_at_k(binary_hit, top)
#             m4 += ndcg_at_k(binary_hit, top)

#             rr = [1/(link_cwe.index(cwe)+1) for cwe in bron_cwe if cwe in link_cwe]
#             m5 += np.mean(rr) if len(rr)>0 else 0
        precision.append(m1/cnt)
        recall.append(m2/cnt)
#         hit.append(m3/cnt)
#         ndcg.append(m4/cnt)
#         mrr.append(m5/cnt)
    return precision, recall # , hit, ndcg, mrr

precision, recall = metric(False, False)
precision_mv, recall_mv = metric(True, False)
precision_ft, recall_ft = metric(False, True)
precision_mv_ft, recall_mv_ft = metric(True, True)
```

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np

def plot(l1, l2, l3, l4, legend1: str, legend2: str, legend3: str, legend4: str):
    labelsize = 20
    plt.subplots(1, 1, figsize = (7, 6))
    matplotlib.rc('xtick',  labelsize=labelsize) 
    matplotlib.rc('ytick',  labelsize=labelsize) 
    matplotlib.rc('axes',   labelsize=labelsize)
    matplotlib.rc('legend', fontsize=13) 

    # ind = np.arange(len(thresholds))
    ind = np.arange(len(n_cves))
    width = 0.2     
    plt.bar(ind,           l1, width, label=legend1)
    plt.bar(ind + width,   l2, width, label=legend2)
    plt.bar(ind + 2*width, l3, width, label=legend3)
    plt.bar(ind + 3*width, l4, width, label=legend4)

    plt.ylabel('Value (0-1)')
    # plt.xlabel('Threshold')
    plt.xlabel('N-CVEs')

    # plt.xticks(ind + 0.5*width, [str(th) for th in thresholds])
    plt.xticks(ind + 1.5*width, [str(th) for th in n_cves])
    plt.legend(loc='lower right')
    plt.show()

plot(precision, precision_mv, precision_ft, precision_mv_ft, 'org', 'multi-view', 'sys context (Linux)', 'multi-view + sys context')
plot(recall, recall_mv, recall_ft, recall_mv_ft,       'org', 'multi-view', 'sys context (Linux)', 'multi-view + sys context')
```
